## Data Cleaning

### Library Import & Initialization

```python
import re
import string
import nltk
import emoji
import contractions
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0  # Make results consistent

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Keep negations & intensifiers
keep_words = {"more", "not", "no", "never", "n't", "very", "too", "so"}
stop_words = {w for w in stop_words if w not in keep_words}
```

### Duplicated Review Removal

```python
# Drop exact duplicates
df = df.drop_duplicates(subset=['review_content'], keep='first').reset_index(drop=True)

# Verify result
print("Total duplicated review: ", df['review_content'].duplicated().sum())
```

### Non-English Review Removal

```python
# Handle non-english review
def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

# Filter dataset
df = df[df['review_content'].apply(is_english)]
df = df.reset_index(drop=True)

# Verify result
df['lang'] = df['review_content'].astype(str).apply(lambda x: detect(x) if x.strip() != '' else '')
df['lang'].value_counts()
```

### Characters Cleaning (Lowercase, Splitting, Removal)

```python
def split_on_capitals(text):
    # Add space before capital letters that are combined with a lowercase letter
    return re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', text)

def clean_text(text):
    if isinstance(text, str):
        text = split_on_capitals(text)
        # Lowercase text
        text = text.lower()
        # Expand contractions before punctuation removal (to cater "isn't" to "is", "not")
        text = contractions.fix(text)
        # Convert emojis to text
        text = emoji.demojize(text, delimiters=(" ", " "))
        # Remove URLs
        text = re.sub(r'https?:\/\/.*?\.(?:jpg|jpeg|png|gif|webp|pdf|docx?)', '', text, flags=re.IGNORECASE)
        text = re.sub(r'https?://\S+|www\.\S+', '', text) # any kinds of url
        # Remove markdown links
        text = re.sub(r'\[.*?\]\(.*?\)', '', text)
        # Remove mentions
        text = re.sub(r'@\w+', '', text)
        # Replace full stops with space
        text = text.replace('.', ' ')
        # Replace commas with space
        text = text.replace(',', ' ')
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        return text
    return text

# Apply preprocessing
df['Cleaned_Review'] = df['review_content'].apply(clean_text)

```

```python
# Check on URLs, mentions and HTML tags
df['has_url'] = df['Cleaned_Review'].str.contains(r'http\S+', case=False)
df['has_mention'] = df['Cleaned_Review'].str.contains(r'@\w+')
df['has_html'] = df['Cleaned_Review'].str.contains(r'<[^>]+>')
df[['has_url', 'has_mention', 'has_html']].sum()
```

### Tokenization

```python
# Tokenize
def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text)
    return text

# Apply tokenization
df['Tokenized_Review'] = df['Cleaned_Review'].apply(tokenize_text)
```

### Stop words Removal

```python
# Remove stopwords
def remove_stopwords(tokens):
    if isinstance(tokens, list):
        return [word for word in tokens if word not in stop_words]
    return tokens

# Apply stopword removal
df['No_Stopwords_Review'] = df['Tokenized_Review'].apply(remove_stopwords)
```

### POS Tagging & Lemmatization

```python
# Map POS tags to WordNet format
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Lemmatize with POS tagging
def lemmatize_tokens(tokens):
    if isinstance(tokens, list):
        pos_tags = nltk.pos_tag(tokens)
        return [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags]
    return tokens

# Apply lemmatization
df['Lemmatized_Review'] = df['No_Stopwords_Review'].apply(lemmatize_tokens)
```

### Feature Extraction (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Join lemmatized tokens back into strings
df['Processed_Text'] = df['Lemmatized_Review'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')

# Initialize TF-IDF vectorizer
tfidf = TfidfVectorizer(
    max_features=5000,        
    ngram_range=(1,2),        # Capture important words
    min_df=2,                 # Ignore rare words
    max_df=0.85,              # Ignore common words
    lowercase=False         
)

# Fit & transform
X_tfidf = tfidf.fit_transform(df['Processed_Text'])

# Show sample features
print("TF-IDF shape:", X_tfidf.shape) 

feature_names = tfidf.get_feature_names_out()
print("Sample features:", feature_names[:20])
```

## Feature Labelling (For Supervised Learning & EDA)

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

def get_sentiment_scores(text):
    if isinstance(text, str):
        return analyzer.polarity_scores(text)

def categorize_sentiment(compound_score):
    if compound_score >= 0.05:
        return "Positive"
    elif compound_score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment scoring
df['Sentiment_Scores'] = df['Cleaned_Review'].apply(get_sentiment_scores)

# Extract individual polarity scores
df['Negative'] = df['Sentiment_Scores'].apply(lambda x: x['neg'])
df['Neutral']  = df['Sentiment_Scores'].apply(lambda x: x['neu'])
df['Positive'] = df['Sentiment_Scores'].apply(lambda x: x['pos'])
df['Compound'] = df['Sentiment_Scores'].apply(lambda x: x['compound'])

# Assign sentiment categories
df['Sentiment_Label'] = df['Compound'].apply(categorize_sentiment)

# Display Result
df[['Cleaned_Review', 'Negative', 'Neutral', 'Positive', 'Compound', 'Sentiment_Label']].head()
```

## Data Sampling

```python
# Sampling using smote to address imbalance issue
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Join lemmatized tokens back into strings
df['Processed_Text'] = df['Lemmatized_Review'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')

# TF-IDF Vectorization
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_tfidf = tfidf.fit_transform(df['Processed_Text'])

# Target
y = df['Sentiment_Label']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, stratify=y, random_state=42
)

# Apply SMOTE only on training data
smote = SMOTE(k_neighbors=1, random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Before SMOTE: {y_train.value_counts()}")
print(f"After SMOTE: {y_train_res.value_counts()}")
```

## Data Handling Over

```python
import joblib
import scipy.sparse as sp
import pandas as pd
import numpy as np
import nlpaug.augmenter.word as naw
import nltk
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from tqdm import tqdm

# Download NLTK data for synonyms
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize Augmenter (Synonym Replacement)
aug = naw.SynonymAug(aug_src='wordnet', aug_p=0.3)

# A. Prepare dataset for Deep Learning
# Initial Split
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(
    df["Cleaned_Review"],
    df["Sentiment_Label"],
    test_size=0.2,
    stratify=df["Sentiment_Label"],
    random_state=42
)

# Combine for easier processing
train_data = pd.DataFrame({'text': X_train_dl, 'label': y_train_dl})

print("Balancing Data via Synonym Augmentation...")

# 1. Separate Classes
df_majority = train_data[train_data.label == 'Positive'] 
df_minority = train_data[train_data.label != 'Positive'] 

# 2. Calculate Gap
majority_count = len(df_majority)
minority_count = len(df_minority)
samples_needed = majority_count - minority_count

print(f"   -> Majority (Pos): {majority_count}")
print(f"   -> Minority (Neg/Neu): {minority_count}")
print(f"   -> Generating {samples_needed} new augmented samples...")

# 3. Generate Augmented Samples
if samples_needed > 0:
    to_augment = resample(df_minority, n_samples=samples_needed, random_state=42)

    new_texts = []

    for text in tqdm(to_augment['text'].tolist(), desc="Augmenting"):
        # Generate augmented text
        augmented_text = aug.augment(text)[0]
        new_texts.append(augmented_text)

    # Create DataFrame for new samples
    df_augmented = pd.DataFrame({
        'text': new_texts,
        'label': to_augment['label'].values # Keep original labels
    })
    # 4. Combine Original + Augmented
    df_balanced = pd.concat([df_majority, df_minority, df_augmented])
else:
    df_balanced = train_data

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
print(f"-> Final Balanced Train Size: {len(df_balanced)}")

# Save DL-ready datasets
df_balanced.to_csv("data/train_dl.csv", index=False)
pd.DataFrame({
    "text": X_test_dl,
    "label": y_test_dl
}).to_csv("data/test_dl.csv", index=False)

# B. Save TF-IDF + SMOTE for Traditional ML
joblib.dump(tfidf, "features/tfidf_vectorizer.pkl")

sp.save_npz("data/train_tfidf.npz", X_train_res)
sp.save_npz("data/test_tfidf.npz", X_test)

y_train_res.to_csv("data/y_train.csv", index=False)
y_test.to_csv("data/y_test.csv", index=False)

print("Preprocessing completed: ML & DL datasets saved.")
```
